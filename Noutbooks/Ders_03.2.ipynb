```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 1. Məlumatları hazırlayırıq
# X - Yürüş (min km), Y - Qiymət (min AZN)
X = np.array([10, 50, 100, 150, 200]).reshape(-1, 1) # Giriş mütləq 2D olmalıdır
Y = np.array([25, 20, 15, 12, 8])                  # Hədəf dəyərlər

# 2. Modeli yaradırıq və təlim edirik
model = LinearRegression()
model.fit(X, Y) # Model X və Y arasındakı riyazi əlaqəni öyrənir

# 3. Yeni bir avtomobil üçün proqnoz veririk
# Məsələn, 120 min km yürüşü olan maşının qiyməti neçə olar?
yeni_yurus = np.array([[120]]) 
proqnoz = model.predict(yeni_yurus)

print(f"120 min km yürüşü olan avtomobilin təxmini qiyməti: {proqnoz[0]:.2f} min AZN")

# 4. Vizual olaraq görək (Matplotlib)
plt.scatter(X, Y, color='blue', label='Mövcud məlumatlar') # Real nöqtələr
plt.plot(X, model.predict(X), color='red', label='Proqnoz xətti') # Modelin çəkdiyi xətt
plt.xlabel('Yürüş (min km)')
plt.ylabel('Qiymət (min AZN)')
plt.legend()
plt.show()
```
