```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 1. Məlumatların hazırlanması
# X - Oxuma saatı (saat), Y - İmtahan balı (0-100)
# Qeyd: .reshape(-1, 1) modelin sütun şəklində məlumat gözləməsi üçün vacibdir.
X = np.array([2, 5, 8, 10, 15, 20]).reshape(-1, 1) 
Y = np.array([30, 45, 60, 65, 85, 95])

# 2. Modelin yaradılması və təlim edilməsi (Fitting)
model = LinearRegression()
model.fit(X, Y) # Model saat və bal arasındakı asılılığı öyrənir

# 3. Yeni məlumat üçün proqnoz (Məsələn: 12 saat oxuyan tələbə)
yeni_saat = np.array([[12]])
proqnoz_bal = model.predict(yeni_saat)

print(f"12 saat oxuyan tələbənin gözlənilən balı: {proqnoz_bal[0]:.2f}")

# 4. Nəticələrin vizuallaşdırılması
plt.scatter(X, Y, color='blue', label='Real Tələbələr') # Real nöqtələr
plt.plot(X, model.predict(X), color='red', label='Proqnoz Xətti') # Öyrənilmiş xətt
plt.title('Oxuma Saatı və İmtahan Balı Arasındakı Əlaqə')
plt.xlabel('Həftəlik Oxuma Saatı')
plt.ylabel('İmtahan Balı')
plt.legend()
plt.grid(True)
plt.show()
```
